### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the number of releases in a year as a parameter of popularity of a genre

Hint 

1: you need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa.
Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 2: Use groupby on the relevant column and use sum() on the same to find out the nuumber of releases in a year/genre.  

```python
#Your answer here
pd.set_option('display.max_columns', 100)
df_items_year=item.copy()
df_items_year['release year']=df_items_year['release date'].apply(lambda x: x.split('-')[2])
df_items_grpby_year=df_items_year.groupby(['release year']).sum()
df_items_grpby_year=df_items_grpby_year.drop(['movie id'], axis=1)

df_items_grpby_year.cumsum().plot(kind='line', figsize=(17, 10))
plt.title("Cumulative Sum of movie per Genre Per Year", y=1.013, fontsize=22)
plt.xlabel("Year", labelpad=16)
plt.ylabel("Count", labelpad=16);

```

```python
figure, axes = plt.subplots(figsize=(10, 20))
sns.heatmap(df_items_grpby_year, annot=True, fmt="d", linewidths=.5, ax=axes)
axes.axes.set_title("Count of movies per Genre by Year", fontsize=20, y=1.01)
axes.axes.set_ylabel("Year", labelpad=50, rotation=0)
axes.axes.set_xlabel("Genre", labelpad=16);
plt.yticks(rotation=0);
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your answer here
df_data=data.copy()

df_all_movie_mean=df_data.groupby(['movie id']).mean()
df_all_movie_mean=df_all_movie_mean.round(2) 
df_all_movie_mean.reset_index(inplace=True)
df_all_movie_mean=df_all_movie_mean[['movie id','rating']]
df_all_movie_mean.columns=['movie id',"rating_avg"]
#
df_movie_100_review=df_data.groupby(['movie id']).count()
df_movie_100_review=df_movie_100_review[['user id']]
df_movie_100_review.columns=["rating_count"]
df_movie_100_review=df_movie_100_review[df_movie_100_review.rating_count>100]
df_movie_100_review.reset_index(inplace=True)
df_merged=df_movie_100_review[['movie id','rating_count']].merge(df_all_movie_mean[['movie id','rating_avg']],on='movie id').merge(item[['movie id','movie title']],on='movie id')

df_merged.nlargest(25,'rating_avg')[['movie id','movie title','rating_count','rating_avg']]
```

### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women


1. There is no need to conduct statistical tests around this. Just compare the percentages and comment on the validity of the above statements.

2. you might want ot use the .sum(), .div() function here.
3. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

#### Conclusion:



```python
df_all_merged=user[['user id','gender']].merge(data[['user id','movie id']],on='user id').merge(item,on='movie id')

cols_list = list(df_all_merged)
cols_list.remove("user id")
cols_list.remove("movie id")
cols_list.remove("unknown")
df_all_grpby=df_all_merged[cols_list].groupby(['gender']).sum()
df_all_grpby
```

```python
df_all_cnt=df_all_grpby.sum()
df_percent = df_all_grpby.div(df_all_cnt)*100
df_percent.round()
```
